```python
# Generate signals
x = np.r_[:1:0.01] # if you don't specify a number before the colon, the starting index defaults to 0
y1 = np.exp( -x )
y2 = np.sin( x*10.0 )/4.0 + 0.5
import matplotlib.pyplot as plt
%matplotlib inline
plt.figure()
plt.plot( x, y1 )
plt.plot( x, y2 , 'g')
plt.xlabel( "Time" )
plt.ylabel( "Space" )

plt.title( "Most important graph in the world" )

plt.legend( ("blue", "red") )
```
